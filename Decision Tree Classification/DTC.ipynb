### Step 1 : Read the Dataset

```python
import pandas as pd

df = pd.read_csv('iris.csv')

df
```

### Step 2 : Basic data quality checks

```python
df.info()
```

```python
m = df.isna().sum()

m[m>0]
```

```python
df.duplicated().sum()
```

### Step 3 : Seperate X and Y (species)

```python
X = df.drop(columns=['species'])
Y = df[['species']]
```

```python

X.head()
```

```python
Y.head()
```

```python

Y.value_counts()
```

### Step 4: Perform preprocessing on X


```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')), ('scaler', StandardScaler())]).set_output(transform='pandas')
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

### Step 5: Apply train test split

```python

from sklearn.model_selection import train_test_split
xtrain, xtest,ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=42)
```

```python
xtrain.head()
```

```python
xtest.head()
```

```python

ytrain.head()
```

```python

ytest.head()
```

### Step 6: Build the decision tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=1, min_samples_split=5, min_samples_leaf=5, criterion='gini')
model.fit(xtrain, ytrain)
```

```python
model.score(xtrain, ytrain)
```

```python
model.score(xtest, ytest)
```

### Hyperparameter tuning 

```python
params = {
    'max_depth':[1,2,3,4,5],
    'min_samples_split' : [3,4,5,6,7],
    'min_samples_leaf' : [5,6,7,8,9],
    'criterion' : ['gini', 'entropy']
}
```

```python
from sklearn.model_selection import GridSearchCV
dtc = DecisionTreeClassifier()
gscv_dtc = GridSearchCV(estimator=dtc, param_grid=params, cv=5, scoring='f1_macro')
gscv_dtc.fit(xtrain, ytrain)
```

```python
gscv_dtc.best_params_
```

```python
gscv_dtc.best_score_
```

```python
best_dtc = gscv_dtc.best_estimator_
best_dtc
```

```python
best_dtc.score(xtrain, ytrain)
```

```python
best_dtc.score(xtest, ytest)
```

### Model Evaluation 

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_dtc, xtrain, ytrain)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(best_dtc, xtest, ytest)
```

```python
# predict the results for train and test 

ypred_train = best_dtc.predict(xtrain)
ypred_test = best_dtc.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytrain, ypred_train))
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest, ypred_test))
```

```python
from sklearn.model_selection import cross_val_score
scores_f1 = cross_val_score(best_dtc, xtrain, ytrain, cv=5, scoring='f1_macro')
scores_f1
```

```python
scores_f1.mean()
```
