### Step 3: Seperate X and Y()

```python
X= df.drop(columns=['Id', 'SalePrice'])
Y = df[['SalePrice']]
```

```python
Y.head()
```

### Step 4 : Preprocessing on X

```python
cat = list(X.columns[X.dtypes=='object'])
con = list(X.columns[X.dtypes!='object'])
```

```python
cat
```

```python
con
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')), ('scaler', StandardScaler())])
```

```python
cat_pipe1 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant', fill_value='Not_Avail')), ('ordinal', OrdinalEncoder())])
```

```python
pre1 = ColumnTransformer([('num', num_pipe1, con), ('cat', cat_pipe1, cat)]).set_output(transform='pandas')
```

```python
X_pre1 = pre1.fit_transform(X)
X_pre1
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
lr = LinearRegression()
sel = SequentialFeatureSelector(estimator=lr, n_features_to_select='auto', direction='backward')
sel.fit(X_pre1, Y)
sel_cols = sel.get_feature_names_out()
sel_cols
```

```python
len(sel_cols)
```

```python
imp_cols = []
for i in sel_cols:
    s= i.split('__')[1]
    imp_cols.append(s)
```

```python
X_sel = X[imp_cols]
X_sel
```

```python
cat_sel = list(X_sel.columns[X_sel.dtypes=='object'])
con_sel = list(X_sel.columns[X_sel.dtypes!='object'])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
num_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='mean')), ('scaler', StandardScaler())])
```

```python
cat_pipe2 = Pipeline(steps=[('impute', SimpleImputer(strategy='constant', fill_value='Not_avail')), ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
```

```python
pre2 = ColumnTransformer([('num', num_pipe2, con_sel), ('cat', cat_pipe2, cat_sel)]).set_output(transform='pandas')
```

```python
X_sel_pre = pre2.fit_transform(X_sel)
X_sel_pre
```

### Step 5: Perform train and test split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_sel_pre, Y, test_size=0.2, random_state=21)
```

### Step 6: build and evaluate the model

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error, r2_score
```

```python
def evaluate_model(model, x, y):
    ypred = model.predict(x)

    mse = mean_squared_error(y, ypred)
    rmse = mse**(1/2)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)

    # print all the results

    print(f'Mean Squared Error : {mse:.2f}')
    print(f'Root Mean Squared Error : {rmse:.2f}')
    print(f'Mean Absolute Error : {mae:.2f}')
    print(f'Mean Absolute Percentage Error : {mape:.4f}')
    print(f'R2 score : {r2:.4f}')
```

```python
best_ridge
```

```python
best_lasso
```

```python
for i in [model1, best_ridge, best_lasso]:
    print(type(i))
    print(f'Training Results')
    evaluate_model(i, xtrain, ytrain)
    print(f'Testing Results')
    evaluate_model(i, xtest, ytest)
    print('\n=================================================\n')
```
